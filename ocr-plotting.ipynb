```python
import plotnine as p9
import pandas as pd
```

```python
ocr_data = pd.read_csv('ocr-with-html.csv')
```

```python
print(ocr_data.columns)
```

```python
p9.ggplot(data=ocr_data, mapping=p9.aes(x='ignore percentage')) + p9.geom_histogram(binwidth=3)
```

```python
p9.ggplot(data=ocr_data, mapping=p9.aes(x='dictionary percentage (of total)')) + p9.geom_histogram(binwidth=3) + p9.xlim(0,100)
```

```python
p9.ggplot(data=ocr_data, mapping=p9.aes(x='dictionary percentage (of non-ignored)')) + p9.geom_histogram(binwidth=3) + p9.xlim(0,100)
```
